```python
from sklearn import utils
from tqdm import tqdm
tqdm.pandas(desc="progress-bar")

'''The paragraph vectors are obtained by training a neural network on the task of predicting a 
probability distribution of words in a 
paragraph given a randomly-sampled word from the paragraph.'''

'''If dm=0, distributed bag of words (PV-DBOW) is used; if dm=1,‘distributed memory’ (PV-DM) is used.
vector_size = 300- dimensional feature vectors.
min_count=2, ignores all words with total frequency lower than this.
negative=5 , specifies 5 “noise words” should be drawn, such as 'the', 'of' etc.
hs=0 , and negative is non-zero, negative sampling will be used, which will allow us to only modify a small percentage 
of the weights, rather than all of them for each training sample.
sample=0 , the threshold after which higher-frequency words are randomly down sampled.'''

model_dbow = Doc2Vec(dm=0, vector_size=300, negative=5, hs=0, min_count=2, sample = 0) #distributed bag of words
model_dbow.build_vocab([x for x in tqdm(pd.Series(train_tagged).values)])
for epoch in range(50):
    model_dbow.train(utils.shuffle([x for x in tqdm(train_tagged.values)]), total_examples=len(train_tagged.values), epochs=1)
    model_dbow.alpha -= 0.002
    model_dbow.min_alpha = model_dbow.alpha
```
